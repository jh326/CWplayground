```python
def plot_hist_each_column(df):
    plt.rcParams['figure.figsize'] = [20, 16]
    fig = plt.figure(1)

    # df의 column 갯수 만큼의 subplot을 출력합니다.
    for i in range(len(df.columns)):
        ax = fig.add_subplot(5, 5, i+1)
        plt.hist(df[df.columns[i]], bins=50)
        ax.set_title(df.columns[i])
    plt.show()
```

```python
plot_hist_each_column(batter_features_df)
```

```python
pd.options.mode.chained_assignment = None
```

```python
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean)/series_std)
    return df
```

```python
scale_columns = ['경기', '타석', '타수', '안타', '홈런', '득점', '타점', '볼넷',
     '삼진', '도루','사구', '타율', '출루', '장타', 'OPS', 'Woba', 'wRC+',
    'WAR*', 'WPA']
batter_df = standard_scaling(batter, scale_columns)
```

```python
batter_df = batter_df.rename(columns={'타율': 'y'})
batter_df.head(5)
```

```python
team_encoding = pd.get_dummies(batter_df['팀'])
batter_df = batter_df.drop('팀', axis=1)
batter_df = batter_df.join(team_encoding)
```
